```python
import os
import matplotlib.pyplot as plt
import torchvision
import torch
from torchvision import datasets, transforms
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img=Image.open(r"C:\Users\Andre\OneDrive - Danmarks Tekniske Universitet\MasterThesis\concept_bottleneck\data\CUB_200_2011\images\087.Mallard\Mallard_0089_77068.jpg").convert('RGB')
plt.imshow(img)
```

```python
transform = transforms.Compose([
    transforms.CenterCrop(299),
    transforms.ToTensor(), #implicitly divides by 255
    ])

plt.imshow(transform(img).permute(1, 2, 0))
```

```python
transform = transforms.Compose([
    transforms.Resize((299, 299)),
    transforms.ToTensor(), #implicitly divides by 255
    ])

plt.imshow(transform(img).permute(1, 2, 0))
```

```python
#set torch random seed
torch.manual_seed(4)
transform = transforms.Compose([
    transforms.RandomResizedCrop(299),
    transforms.ToTensor(), #implicitly divides by 255
    ])

fig, axs = plt.subplots(1,4)
for i in range(4):
    axs[i].imshow(transform(img).permute(1, 2, 0))
    axs[i].axis('off')

plt.show()
```
